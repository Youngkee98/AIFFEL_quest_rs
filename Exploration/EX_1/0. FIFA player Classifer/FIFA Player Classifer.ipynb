```python
import pandas as pd

# 같은 디렉토리에 있으니 파일명만!
file_path = "Career Mode female player datasets - FIFA 16-22.csv"

# 시트 이름 확인
original_player = pd.read_csv(file_path, encoding='cp1252')
original_player.head()


```

```python

fifa_22 = original_player.copy() # 복사본을 만들어준다.
fifa_22.dropna(axis=1, how='all', inplace=True) # axis : 0은 행, 1은 열, how : any는 하나라도 있으면 drop, all은 모두가 null이어야 drop
fifa_22.reset_index(drop=True, inplace=True) # 인덱스를 리셋해준다.


print(original_player.shape)
print('전처리된 행,전처리된 열: ',end='')
print(fifa_22.shape) # 22개의 열이 삭제되었다.
```
